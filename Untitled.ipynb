```python
by_env = gym.make('by-v0')
eckel_multienv = gym.make('eckel-mdisc-v0')
eckel_env = gym.make('eckel-disc-v0')
```

```python
dqnModel = DQN(MlpPolicy, eckel_env, verbose = 1)
dqnModel.learn(total_timesteps = 75000, log_interval = 1)
mean_reward, std_reward = evaluate_policy(dqnModel, dqnModel.get_env(), n_eval_episodes=10)
dqnModel.save("notebook_demo")
print(mean_reward)
print(std_reward)
```

```python
obs = eckel_env.reset()
action_list = []
counter = 0
for i in range(0,50):
    action, _states = dqnModel.predict(obs, deterministic = True)
    bs, reward, done, info = eckel_env.step(action)
    action_list.append(action)
    eckel_env.render()

print(action_list)
    
```
